```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv('Customer_Support_Tickets.csv')
```

```python
# Step 2: Data Cleaning and Preprocessing

# Check for missing values
print(df.isnull().sum())

df.dropna(subset=['Resolution', 'Time to Resolution', 'Customer Satisfaction Rating'], inplace=True)
```

```python
# Convert "First Response Time" to datetime
df['First Response Time'] = pd.to_datetime(df['First Response Time'], errors='coerce')

# Impute missing values in "First Response Time" with the median
median_response_time = df['First Response Time'].median()
df['First Response Time'].fillna(median_response_time, inplace=True)

# Check if missing values are filled
print(df['First Response Time'].isnull().sum())
```

```python
# Convert "Time to Resolution" to datetime
df['Time to Resolution'] = pd.to_datetime(df['Time to Resolution'], errors='coerce')

# Impute missing values in "Time to Resolution" with the median
median_resolution_time = df['Time to Resolution'].median()
df['Time to Resolution'].fillna(median_resolution_time, inplace=True)

# Check if missing values are filled
print(df['Time to Resolution'].isnull().sum())
```

```python
# Impute missing values in "Customer Satisfaction Rating" with the median
median_rating = df['Customer Satisfaction Rating'].median()
df['Customer Satisfaction Rating'].fillna(median_rating, inplace=True)
print(df['Customer Satisfaction Rating'].isnull().sum())
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

# Print summary statistics of the dataframe
print(df.describe())

# Print correlation matrix of the dataframe
print(df.corr())

# Create a pairplot of the dataframe
sns.pairplot(df)
plt.show()
```
